```python
from os import replace
#Task 2.2: Create and Fix Missing Value
# Artificially create missing data for practice
import numpy as np

# Your tasks:
# 1. Make a copy of the dataset: df_practice = df.copy()
# 2. Randomly set 50 math scores to np.nan
# 3. Randomly set 30 writing scores to np.nan
# 4. Fill missing math scores with the MEDIAN math score
# 5. Fill missing writing scores with the MEAN writing score
# 6. Verify no missing values remain

# Steps to follow:
# Step 1: Create copy and show original missing values
df_practice = df.copy()
print(f"Original missing values:\n{df_practice.isnull().sum()}\n")


# Step 2: Create missing data
np.random.seed(20)  # For reproducible results
missing_math_idx = np.random.choice(df_practice.index, 50, replace = False)
#missing_reading_idx = np.random.choice(df_practice.index, 35, replace = False)
missing_writing_idx = np.random.choice(df_practice.index, 30, replace = False)

# Step 3: Repalce missing data with Nan Values
df_practice.loc[missing_math_idx, 'math score'] = np.nan
#df_practice.loc[missing_reading_idx, 'reading score'] = np.nan
df_practice.loc[missing_writing_idx, 'writing score'] = np.nan

# Step 4: Repalce missing data with Nan Values
print(f"Missing values after modified:\n{df_practice.isnull().sum()}\n")

# Step 5: Find the MEDIAN math score and MEAN writing score
median_math = df_practice['math score'].median()
mean_writing = df_practice['writing score'].mean()

print(f"The MEDIAN math score: {round(median_math, 1)}")
print(f"The MEAN writing score: {round(mean_writing, 1)}\n")

# Step 5: Fill missing data with the MEDIAN math score and MEAN writing score
df_practice['math score'] = df_practice['math score'].fillna(median_math)
df_practice['writing score'] = df_practice['writing score'].fillna(mean_writing)

# Step 6: Final check - should be 0 missing values now
print(f'Missing values after replacement:\n{df_practice.isnull().sum()}')

# Your code here:
```

```python
#Task 3.3: Use your data analysis skills to answer these easy questions:

# 1. Who performs better in math - males or females?
#    - Calculate average math score by gender
avg_math_score_by_gender = df.groupby('gender')['math score'].mean()
print(f'average math score by gender:\n {avg_math_score_by_gender}')
print('\n')
#    - Show the difference between male and female average
difference = avg_math_score_by_gender['male'] - avg_math_score_by_gender['female']
print(f'the difference between male and female average: {difference:.2f}')
print('\n')

# 2. Which subject do students perform best in overall?
#    - Calculate the overall average for math, reading, and writing
overall_average = df[['math score', 'reading score', 'writing score']].mean().mean()
print(f'the overall average for math, reading, and writing: {overall_average:.2f}')
print('\n')

#    - Rank the subjects from highest to lowest average score
rank_subject = df[['math score', 'reading score', 'writing score']].mean().sort_values(ascending = False)
print(f'Rank the subjects from highest to lowest average score:\n{rank_subject}\n')

# 3. What's the impact of parental education?
#    - Find the average total score for each parental education level
df['total score'] = df['math score'] + df['reading score'] + df['writing score']
avg_total_score_by_parent = df.groupby('parental level of education')['total score'].mean()
print(f'the average total score for each parental education level:\n{avg_total_score_by_parent}\n')

#    - Identify which education level leads to highest student performance
highest_performance = avg_total_score_by_parent.idxmax()
highest_score = avg_total_score_by_parent.max()
print(f'{highest_performance} education level leads to highest student performance with score {highest_score:.2f}')
```
